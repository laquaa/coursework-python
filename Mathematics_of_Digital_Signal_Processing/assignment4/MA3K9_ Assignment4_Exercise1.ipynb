```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import freqz

# Coefficients of the transfer function
# Numerator coefficients (1 - 0.5z^-1)
numerator = [1, -0.5]
# Denominator coefficients (1 - 0.9z^-1)
denominator = [1, -0.9]

# Compute the frequency response
# 'freqz' computes the frequency response of a digital filter
w, h = freqz(numerator, denominator, worN=1024)  # worN is the number of frequency points

# Calculate magnitude and phase
magnitude = np.abs(h)  # Magnitude response
phase = np.angle(h)    # Phase response

# Convert frequency to normalized frequency (0 to )
normalized_frequency = w / np.pi

# Plot the magnitude response
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(normalized_frequency, 20 * np.log10(magnitude))  # Magnitude in dB
plt.title('Frequency Response of H(z)')
plt.ylabel('Magnitude (dB)')
plt.grid()
plt.xlabel('Normalized Frequency ( rad/sample)')

# Plot the phase response
plt.subplot(2, 1, 2)
plt.plot(normalized_frequency, np.unwrap(phase))  # Unwrap the phase to avoid discontinuities
plt.ylabel('Phase (radians)')
plt.grid()
plt.xlabel('Normalized Frequency ( rad/sample)')

# Show the plots
plt.tight_layout()
plt.show()
```
