```python
# 2 (b)

import numpy as np
import matplotlib.pyplot as plt

# Define constants
gamma = np.pi / 4
a = 0.9
c = np.pi / 3

# Define frequency range 
omega = np.linspace(-np.pi, np.pi, 1000)

# Calculate Z(e^j)
Z = (np.exp(1j * gamma) / (2 * (1 - a * np.exp(1j * (c - omega))))) + \
    (np.exp(-1j * gamma) / (2 * (1 - a * np.exp(-1j * (c + omega)))))

# Compute magnitude and phase spectra
magnitude_Z = np.abs(Z)  # Magnitude spectrum
phase_Z = np.angle(Z)    # Phase spectrum

# Plot magnitude spectrum
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(omega, magnitude_Z)
plt.title('Magnitude Spectrum |Z(e^j)|')
plt.xlabel('Frequency ()')
plt.ylabel('Magnitude')
plt.grid(True)

# Plot phase spectrum
plt.subplot(2, 1, 2)
plt.plot(omega, phase_Z)
plt.title('Phase Spectrum Z(e^j)')
plt.xlabel('Frequency ()')
plt.ylabel('Phase (radians)')
plt.grid(True)

# Show plots
plt.tight_layout()
plt.show()
```
