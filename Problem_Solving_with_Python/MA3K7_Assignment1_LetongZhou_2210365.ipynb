```python
def generate_number_bracelet(a1, a2, length=100):

    # Initialize the sequence with the first two numbers
    sequence = [a1, a2]

    # Generate the sequence up to the desired length
    for _ in range(2, length):
        next_number = (sequence[-1] + sequence[-2]) % 10
        sequence.append(next_number)

    # Format the sequence as "1→5→6→..."
    formatted_sequence = "→".join(map(str, sequence)) + "→"

    return formatted_sequence

# Starting numbers
first_number = 1
second_number = 5

# Generate the sequence
result = generate_number_bracelet(first_number, second_number)

# Print the formatted sequence
print("The first 40 numbers in the sequence are:", result)
```
