```python
#1.7

def x_plus(p,q):
    x_plus = -p + (p**2 + q)**0.5
    return x_plus
def x_minus(p,q):
    x_minus = -p - (p**2 + q)**0.5
    return x_minus
```

```python
#(a)

def R_plus(p,q,delta_p):
    x_p = x_plus(p, q)
    x_p_delta = x_plus(p + delta_p, q)
    R_p = abs((x_p_delta - x_p) / x_p) / abs(delta_p / p)
    return R_p
def R_minus(p,q,delta_p):
    x_m = x_minus(p, q)
    x_m_delta = x_minus(p + delta_p, q)
    R_m = abs((x_m_delta - x_m) / x_m) / abs(delta_p / p)
    return R_m

p_values = [-1, 1]
q = 1
delta_p = 10**(-3)
R_plus_values = []
R_minus_values = []

for p in p_values:
    R_plus_values.append(R_plus(p,q,delta_p))
    R_minus_values.append(R_minus(p,q,delta_p))

R_plus_values, R_minus_values
```

```python
# The codes above calculate R+ when p=-1 & when p=1 seperately and add them to the first list.Then R- when p=-1 & when p=1 are calculated seperately and added to the second list.
# The results above show that the sensitivity of the problem is almost the same for x+ and x-, taking p = 1 and p = -1.
# This possibly means that sensitivity does not depend on whether one considers x+ or xâˆ’ as well as the sign of p.
```

```python
#(b)

import matplotlib.pyplot as plt
import numpy as np

def K_p(p,q):
    condition_number = abs(p/(p**2+q)**0.5)
    return condition_number

p_domain = np.linspace(-1, 1, 200)
condition_number_p = K_p(p_domain,q)
R_plus_results = R_plus(p_domain,q,delta_p)
R_minus_results = R_minus(p_domain,q,delta_p)

plt.subplot(1, 2, 1)
plt.plot(p_domain,R_plus_results,label='R+(p, Î´p)',color='red')
plt.plot(p_domain,condition_number_p,label='K(p)',color='blue')
plt.title('R+(p, Î´p)&K(p) on [-1, 1]')
plt.xlabel('p')
plt.ylabel('R+(p, Î´p)&K(p)')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(p_domain,R_minus_results,label='R-(p, Î´p)',color='red')
plt.plot(p_domain,condition_number_p,label='K(p)',color='blue')
plt.title('R-(p, Î´p)&K(p) on [-1, 1]')
plt.xlabel('p')
plt.ylabel('R-(p, Î´p)&K(p)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#1.8

def linear_interpolate(f,x_0,x_1,x):
    result = f(x_0) + ((f(x_1)-f(x_0))/(x_1-x_0))*(x-x_0)
    return result
```

```python
#(b)

def lagrange_interp(x,nodes,values):
    pm_x = 0
    for k in range(len(nodes)):
        pm_x += lagrange(k,x,nodes)*values[k]
    return pm_x

def f(x):
    result = 1/(1+x**2)
    return result

m_values = [3,5,11,21]
x = np.linspace(-5,5,1000)
for m in m_values:
    nodes = np.linspace(-5,5,m)
    values = f(nodes)
    pm_x = lagrange_interp(x,nodes,values)
    plt.plot(x,pm_x,label='pm(x) m='+str(m))
plt.plot(x,f(x),label='f(x)')
plt.title('f(x)&pm(x) vs. x')
plt.xlabel('x')
plt.ylabel('f(x)&pm(x)')
plt.legend()
plt.show()
```

```python
# As the number of nodes & values used in Lagrange interp increases, the interpolation polynomial better approximates the function f(x) over most of the interval.
# However, near the edges of the interval, especially closed to the boundaries, the interpolant results has severe oscillations, which is known as Runge's phenomenon.
```

```python
x_values = [1,4.5]
all_error = []
for m in m_values:
    errors = []
    for x in x_values:
        nodes = np.linspace(-5,5,m)
        values = f(nodes)
        pm_x = lagrange_interp(x,nodes,values)
        error = abs(pm_x-f(x))
        errors.append(error)
    all_error.append(errors)
all_error
```

```python
# The list above contains the absolute error between the values of interpolants and the actual value of the funvtion.
# The list has 4 sub-lists, corresponding to m=3,5,11,21.
# In each sub-list, the first element is the error when x=1 and the second one is the error when x=4.5.
# The results show that as the number of nodes increases, the interpolation at x=1 becomes exact for 11 and 21 nodes, with the absolute error reaching zero.
# These observations are consistent with the Weierstrass Approximation Theorem, which states that a continuous function can be approximated as closely as desired by a polynomial interpolant given a sufficient number of nodes.
```
